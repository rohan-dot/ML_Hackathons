```python
def check_value_counts_perc(train, test, col):
    result = pd.concat([(train[col].value_counts() / train.shape[0]).reset_index(), 
                        (test[col].value_counts() / test.shape[0]).reset_index()], 
                       axis=1)
    result.columns = ['train_value', 'train_perc', 'test_value', 'test_perc']
    
    return result
```

```python
check_value_counts_perc(train, test, "feature_7")
```

```python

def get_closest(train, col, value):
    a = train[col].unique().tolist()
    min_idx = min(range(len(a)), key=lambda i: abs(a[i]-value))

    return a[min_idx]
```

```python
dont_consider_feats = ['feature_7', 'feature_8', 'feature_9', 'feature_10', 'feature_11', 'feature_12', 'feature_13', 'feature_14','feature_15','labels']

for col in train.columns.tolist():
    if col not in dont_consider_feats:
        
        test[col] = test[col].apply(lambda x: get_closest(train, col, x))
```

```python
numeric_feats = all_data.dtypes[all_data.dtypes != "uint8"].index

# Check the skew of all numerical features
skewed_feats = all_data[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(500)
```

```python
skewness = skewness[abs(skewness) > 0.75]
print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    #all_data[feat] += 1
    all_data[feat] = boxcox1p(all_data[feat], lam)
```

```python

train = all_data[:ntrain]
test = all_data[ntrain:]
```

```python
from sklearn.model_selection import train_test_split, cross_val_score

X_train, X_test, y_train, y_test = train_test_split(
    train, y_train, test_size=0.1, random_state=42
)
```

```python
import numpy as np
X_train[:] = np.nan_to_num(X_train)
```

```python
X_test[:] = np.nan_to_num(X_test)
```

```python
params = {'colsample_bytree': 0.8,'learning_rate': 0.0003, 'max_depth': 8, 'n_estimators': 1000, 'subsample': 1,'scale_pos_weight':6,'objective':'binary:logistic','eval_metric':'logloss'}
```

```python
import xgboost as xgb
dt =xgdmat=xgb.DMatrix(X_train,y_train)
dte = xgdmat=xgb.DMatrix(X_test,y_test)
```

```python
model = xgb.train(params, dt, 1000, [(dt, "train"),(dte, "valid")], verbose_eval=200)
```

```python
from termcolor import colored
#from sklearn.cross_validation import KFold

from sklearn.metrics import confusion_matrix

y_pred1 = model.predict(dte)
# Making the Confusion Matrix
cm = confusion_matrix(y_test, (y_pred1>0.5))
print(colored('The Confusion Matrix is: ', 'red'),'\n', cm)
# Calculate the accuracy on test set
predict_accuracy_on_test_set = (cm[0,0] + cm[1,1])/(cm[0,0] + cm[1,1]+cm[1,0] + cm[0,1])
print(colored('The Accuracy on Test Set is: ', 'blue'), colored(predict_accuracy_on_test_set, 'blue'))
```
